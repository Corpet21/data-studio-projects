```python
import pandas as pd
from bs4 import BeautifulSoup
from html.parser import HTMLParser
import re
import lxml
import datetime
```

```python
doc = BeautifulSoup(open("Bernie Sanders - Events.htm", "r").read())
```

```python
soup = doc.find_all(class_="_24er")
rows = []

for s in soup:
    row = s.text.strip()
    rows.append(row)
```

```python
df = pd.DataFrame(data = rows, columns = ["details"])
```

```python
pd.set_option('display.max_colwidth', -1)
```

```python
df["date"] = df['details'].str.extract(r'(\w\w\w\d\d?)', expand=True)
```

```python
df["description"] = df['details'].str.extract(r'(\b .* · )', expand=True)
```

```python
df["guests"] = df['details'].str.extract(r'(· .* guests)', expand=True)
```

```python
df["state"] = df['details'].str.extract(r'(, [A-Z][A-Z])', expand=True)
```

```python
df["state"] = df.state.str.replace(",","")
```

```python
df["guests"] = df["guests"].str.replace("·", "")
df["guests"] = df["guests"].str.replace("guests", "")
df["guests"] = df["guests"].str.replace(",","")
```

```python
df["guests"] = df["guests"].astype(int)
```

```python
regex_pat = re.compile(r'[A-Z]\w\w\w?\w? \d.*')
df["description"] = df["description"].str.replace(regex_pat,"")
```

```python
regex_pat = re.compile(r'Tomorrow .*')
df["description"] = df["description"].str.replace(regex_pat,"")
```

```python
df["location"] = df["details"].str.extract(r'(guests.*[A-Z][A-Z])')
df["location"] = df["location"].str.replace("guests", "")
```

```python
regex_pat = re.compile(r'\w\w\w')
df["date"].str.replace(regex_pat,"")
```

```python
df = df.drop(("details"), axis=1)
```

```python
df["month"] = df["date"].str.extract(r'(\w\w\w)')
```

```python
df["day"] = df["date"].str.extract(r'(\d\d?)')
```

```python
df["date"] = df['month'] +' '  + df['day'] 
```

```python
df = df.drop(columns=['month','day'])
```

```python
df = df.drop(df.index[82:])
```

```python
df["location"] = df["location"].str.replace("(", ",")
```

```python
df["location"] = df["location"].str.replace(")", ",")
```
